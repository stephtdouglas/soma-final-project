My goal here is to test the difference between having Kepler-like data vs. an irregularly-sampled ground-based survey. Kepler is pretty much the greatest time-series data we have for stars (except maybe for very bright stars with long-term light curves).  Ground-based surveys are less precise (at least for fainter stars) _and_ they are restricted to night-time observations if nothing else. In a non-dedicated survey like PTF, there may only be one datapoint per night.  I'm curious how the outcome changes if we pretend the Kepler light-curve contains the true stellar rotation period, and just sample that light curve at a similar rate to ground-based surveys and measure the period from that.

### Stitch Light Curves Together, then take an irregular sampling


I did not interpolate between the quarters here (though data-downlink gaps within quarters are filled)

```python
full_times = [np.append(np.append(times[k][0],times[k][1]),np.append(times[k][2],times[k][3])) for k in range(len(KIDs))]
full_fluxes = [np.append(np.append(fluxes[k][0],fluxes[k][1]),np.append(fluxes[k][2],fluxes[k][3])) for k in range(len(KIDs))]
full_errors = [np.append(np.append(errors[k][0],errors[k][1]),np.append(errors[k][2],errors[k][3])) for k in range(len(KIDs))]
```

Select a random sampling time, vaguely based on the PTF sampling rate in Praesepe (Agüeros et al. 2011): 185 observations, that (based mostly on eyeballing Figure 3) cover about 100 days, with about half the points concentrated over 5 nights

```python
def random_ptf(times):
    
    cadence = np.median(np.diff(times))
    days5 = 5.0/cadence
    
    # first, only nights can be observed, so only use half the points
    nights = np.where((times-np.floor(times))>0.5)[0]
    # then, datapoints are at minimum 15 minutes apart, but the cadence is 30 min so it's fine!
    
    # We want a 100 day stretch, so the maximum starting time is 100 days before the end of the data
    max_time = times[nights[-1]] - 100
    max_nights = np.where(times[nights]<max_time)[0][-1]
    #print max_nights,times[nights][max_nights]

    # Then choose a 100 day stretch
    sub100 = []
    while len(sub100)==0:
        # Choose the starting night at random (it's really an index of the nights array)
        start_night = np.random.random_integers(0,high=max_nights)

        # We want a 100 day stretch, so the maximum time is 100 days after the starting night
        max_subtime = times[nights[start_night]]+100
        end_night = max(np.where(times[nights]<max_subtime)[0])
        #print "start_night,end_night,max_subtime"
        #print start_night,end_night,max_subtime
        #print "times_start,times_end"
        #print times[nights[start_night]],times[nights[end_night]]

        # Then actually determine the indices that correspond to the times between start_night and end_night
        sub_loc = np.where((nights>=start_night) & (nights<end_night))[0]
        sub100 = nights[nights[sub_loc]]

    #print times[sub100][:10]

    # Take a random sampling of those nights
    sparse_times = sub100[np.random.random_integers(0,high=len(sub100)-1,size=95)]

    # Take a denser sampling over 5 nights
    sub_high = []
    while len(sub_high)==0:
        high_cadence_start = np.random.random_integers(0,max(sub100)-days5)
        sub_high = sub100[(sub100>=high_cadence_start) & (sub100<=(high_cadence_start+days5))]
    high_cadence = sub_high[np.random.random_integers(0,high=len(sub_high)-1,size=90)]
    # Append and sort to return the new times
    new_time_locs = np.sort(np.append(sparse_times,high_cadence))
    
    return new_time_locs
```

The Kepler data isn't flux-calibrated, it's just counts from the telescope. When the telescope is rotated between Quarters, the star sits on different pixels and so the median counts may be higher or lower depending on the sensitivity of the new pixels.  While there are most certaintly data-quality issues with ground-based surveys that I'm not taking into account, I don't think the big jumps are likely, so I'm removing them (mostly).  In a few places, the jumps stay in, I think due to more extreme outliers...

```python
def fix_jumps(flux):
    jump = np.argmax(np.abs(np.diff(flux)))
    std_no_jump = abs(np.std(flux[:jump])+np.std(flux[jump:]))/2.0
    jump_diff = np.abs(np.diff(flux)[jump])
    #print jump,jump_diff,flux[jump-1:jump+2],std_no_jump
    iteration = 1
    while ((jump_diff>std_no_jump) and (iteration<4)):
        norm_change = np.median(flux[:jump])/np.median(flux[jump:])
        #print "norm change", norm_change
        flux[jump:] = flux[jump:] * norm_change
        jump = np.argmax(np.diff(flux))
        std_no_jump = abs(np.std(flux[:jump])+np.std(flux[jump:]))/2.0
        jump_diff = np.abs(np.diff(flux)[jump])
        #print "new",jump,jump_diff, std_no_jump,iteration
        iteration += 1
    return flux
```

```python
def run_random(random_func,times,fluxes,errors,fractional_error=0.001,plot=False):
    """
    calculates and returns a random subset of the observations given by (times, fluxes, errors)
    
    The errors are changed by adding in quadrature the original errors and the fluxes times fractional_error
    
    if plot==True, plots an errorbar plot showing the new points
    
    returns: new_times, new_fluxes, new_errors
    """
    rand_locs = random_func(times)
    
    new_times = times[rand_locs]
    
    new_fluxes = fix_jumps(fluxes[rand_locs])
    
    new_errors = np.sqrt(errors[rand_locs]**2 + (fractional_error*np.median(new_fluxes))**2)
    
    if plot:
        plt.figure()
        plt.errorbar(full_times[0][rand_locs],new_fluxes,new_errors,color='r',fmt=".")
        
    return new_times, new_fluxes, new_errors
```

In this case, I'm assuming the Kepler data uncertainties stay the same.

```python
random_iter = 200
for k,kid in enumerate(KIDs):
    random_periods = np.zeros(random_iter)
    for i in range(random_iter):
        new_times, new_fluxes, new_errors = run_random(random_ptf,full_times[k],full_fluxes[k],full_errors[k],
                                                       fractional_error=0.0,plot=False)
        random_periods[i] = run_ls(new_times, new_fluxes, new_errors,plot=False)
        
    # Then histogram all the resulting periods, and compare with the McQuillan et al. period
    plt.figure(figsize=(16,4))
    ax = plt.subplot2grid((1,6),(0,0),colspan=3)
    ax.hist(random_periods,bins=np.arange(50),color="w")
    ax.set_xlim(0,50)
    ax.set_title(str(kid))
    ax.plot((mcq_periods[k],mcq_periods[k]),ax.get_ylim(),"g:",lw=2,label="Prot={:.2f}".format(mcq_periods[k]))
    med_period = np.median(random_periods)
    ax.plot((med_period,med_period),ax.get_ylim(),"r--",lw=2,label="Prot={:.2f}".format(med_period))
    ax.legend(loc=1)
    ax.set_xlabel("Prot",fontsize="x-large")
    ax.set_ylabel("Number of trials",fontsize="x-large")
    
    ax2 = plt.subplot2grid((1,6),(0,3),colspan=3)
    ax2.errorbar(full_times[k],full_fluxes[k],full_errors[k],color="r",fmt=".")
    ax2.set_xlabel("Time (d)")
    
```

Half- and double-period aliases look like a problem in a few cases (1870952, 1996399, 2142372) - the double-dip in the light-curve is not incredibly obvious in all quarters of data for these stars. Two stars find unusually short periods (1721911, 1866535); either the Kepler data is very noisy or they're undergoing a lot of very rapid variability in most quarters.  And 1866986 shows a whole range of possible periods; it's amplitude of variability changes a lot over the full 4 quarters, and with random sampling it looks like almost any period could be detected.

Now using errors similar to PTF's (about 0.02 mags on R~14 -> 0.1%)

```python
random_iter = 200
for k,kid in enumerate(KIDs):
    random_periods = np.zeros(random_iter)
    for i in range(random_iter):
        new_times, new_fluxes, new_errors = run_random(random_ptf,full_times[k],full_fluxes[k],full_errors[k],
                                                       fractional_error=0.001,plot=False)
        random_periods[i] = run_ls(new_times, new_fluxes, new_errors,plot=False)
        
    # Then histogram all the resulting periods, and compare with the McQuillan et al. period
    plt.figure(figsize=(16,4))
    ax = plt.subplot2grid((1,6),(0,0),colspan=3)
    ax.hist(random_periods,bins=np.arange(50),color="w")
    ax.set_xlim(0,50)
    ax.set_title(str(kid))
    ax.plot((mcq_periods[k],mcq_periods[k]),ax.get_ylim(),"g:",lw=2,label="Prot={:.2f}".format(mcq_periods[k]))
    med_period = np.median(random_periods)
    ax.plot((med_period,med_period),ax.get_ylim(),"r--",lw=2,label="Prot={:.2f}".format(med_period))
    ax.legend(loc=1)
    ax.set_xlabel("Prot",fontsize="x-large")
    ax.set_ylabel("Number of trials",fontsize="x-large")
    
    ax2 = plt.subplot2grid((1,6),(0,3),colspan=3)
    all_new_errors = np.sqrt(full_errors[k]**2 + 0.001*full_fluxes[k])
    ax2.errorbar(full_times[k],full_fluxes[k],all_new_errors,color="r",fmt=".")
    ax2.set_xlabel("Time (d)")
    
```

No change in the results overall with the increase in errors - I'm somewhat surprised. Though maybe re-drawing the flux points from a larger error distribution would have made a bigger difference.

I chose 15 M dwarfs from McQuillan et al. (2013) ([Paper](http://adsabs.harvard.edu/abs/2013MNRAS.432.1203M), [Table](http://www.astro.tau.ac.il/~amy/table2_periodic.txt)) for which those authors detected period variability. All of these stars have 10 d < P<sub>rot</sub> < 40 d.  To match their study, I used the first 4 quarters of Kepler data for these objects, and I used the flux that's been corrected for systematic effects using the Kepler pipeline.

```python
quarters = ["2009259160929","2009350155506","2010078095331","2010265121752"]
mcq = at.read("mcquillan13_table2_cut.txt")
KIDs = mcq["KID"]
mcq_periods = mcq["Period"]

for k,kid in enumerate(KIDs):
    plt.figure(figsize=(15,3))
    for q,quarter in enumerate(quarters):
        ax = plt.subplot(1,4,q+1)
        time,flux,flux_err = open_kepler("lcs/kplr00{}-{}_llc.fits".format(kid,quarter))
        ax.plot(time,flux,"k-")
        ax.set_xlabel("time (d)")
        ax.set_ylabel("flux")
    plt.suptitle(kid)
```

# Patch gaps and remove outliers

Following McQuillan et al., I patched the data gaps (mostly from Kepler turning towards Earth to downlink data) by linearly interpolating between the points on either side of the gap. The functions for this are found in fix_kepler.py

```python
times = []
fluxes = []
errors = []
for k,kid in enumerate(KIDs):
    kid_t = []
    kid_f = []
    kid_e = []
    for q,quarter in enumerate(quarters):
        time,flux,flux_err = open_kepler("lcs/kplr00{}-{}_llc.fits".format(kid,quarter))
        fill_times,fill_fluxes,fill_errs = fill_gaps(time,flux,flux_err)
        kid_t.append(fill_times)
        kid_f.append(fill_fluxes)
        kid_e.append(fill_errs)
    times.append(kid_t)
    fluxes.append(kid_f)
    errors.append(kid_e)
```

# Fast Fourier Transform

Now we can do an FFT on each quarter. I'm calculating on each quarter separately, on the assumption that the period found in McQuillan et al. (2013) is the correct period.

```python
fft_periods = np.zeros(len(KIDs)*len(quarters)).reshape((len(KIDs),len(quarters)))
for k,kid in enumerate(KIDs):
    for q,quarter in enumerate(quarters):
        #print times[k][q],fluxes[k][q]
        fft_periods[k][q] = run_fft(times[k][q],fluxes[k][q],input_period=mcq_periods[k],plot=False)
        #plt.suptitle("{} Q{}".format(kid,q+1))
        
```

Plot the quarter-by-quarter results comparing the FFT period to the McQuillan et al. (2013) period. The solid line indicates a one-to-one correspondence, and the dashed lines indicate double/half-period aliases

```python
plt.figure(figsize=(10,10))
for q,quarter in enumerate(quarters):
    ax = plt.subplot(2,2,q+1)
    q_periods = fft_periods[:,q]
    ax.plot(mcq_periods,q_periods,'ro')
    xx = np.arange(50)
    ax.plot(xx,xx,'k-')
    ax.plot(xx,xx/2,'k--')
    ax.plot(xx,xx*2,'k--')
    ax.set_xlim(0,50)
    ax.set_ylim(0,50)
    ax.set_xlabel("McQuillan et al. (2013) Prot",fontsize="x-large")
    ax.set_ylabel("FFT Prot",fontsize="x-large")
    ax.tick_params(labelsize="large")
    ax.set_title("Quarter {}".format(q+1))
```
